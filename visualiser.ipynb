```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
import scipy.stats as scpstat
```

```python
df = pd.read_csv("Results.csv").to_numpy(dtype=np.float64)
```

```python
# Get x coordinates from file
xearth = []
yearth = []
zearth = []

xj = []
yj = []
zj = []


xsun = []
ysun = []
zsun = []

energy = []
GPE =  []
KE = []
for i in df:
    xearth.append(i[4])
    yearth.append(i[5])
    zearth.append(i[6])

    xj.append(i[12])
    yj.append(i[13])
    zj.append(i[14])

    xsun.append(i[8])
    ysun.append(i[9])
    zsun.append(i[10])
    energy.append(i[0])
    GPE.append(i[1])
    KE.append(i[2])

energy_difference = df[len(df)-1][0] - df[0][0] 
energy_percent_difference = (energy_difference / df[0][0]) * 100
print(df)
print(energy_percent_difference) 
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(xearth,label = "Earth x position")
plt.gca().set_xticks([])
plt.legend(loc = "upper right")
plt.subplot(212)
plt.xlabel("time step")
plt.plot(yearth,label = "Earth y positions")

plt.legend(loc = "upper right")

    
    
```

```python
plt.figure(2)
plt.subplot(211)
plt.plot(xj,label = "Jupiter x position")
plt.gca().set_xticks([])
plt.legend()
plt.subplot(212)
plt.xlabel("time step")
plt.plot(yj,label = "Jupiter y positions")

plt.legend()
```

```python

plt.plot(energy,label = "Total energy")
plt.ylim(0,10e43)

plt.plot(KE,label = "Kinetic Energy")

# plt.plot(KE,label = "KE")
plt.plot(GPE,label = "GPE")
plt.legend()

```

```python
#Percentage change in Energy
temp_E = []

for i in range(0,len(energy)):
    temp = energy[0]
    temp_E.append((-(energy[i] - temp)/temp)*100)

plt.plot(temp_E)
plt.xlabel("time step")
plt.title("Percentage change in Energy over time")
plt.legend()
# print(energy)
```

```python
# Work out distance between earth and sun over time

distance = []

for i in range(0,len(xearth)):
    distance.append(np.sqrt((xearth[i] - xsun[i])**2 + (yearth[i] - ysun[i])**2 + (zearth[i] - zsun[i])**2))
```

```python
plt.plot(distance,label= "Earth Sun Distance")
plt.legend(loc = "upper right")
```

```python
#fractional earth sun distance change
fractional_es_dif = []
initial_r = distance[0]
for i in range(0,len(distance)):
    fractional_es_dif.append((distance[i] - initial_r)/initial_r)
plt.plot(fractional_es_dif)
plt.title("Fractional Change in Earth Sun Distance")
# plt.legend(loc = "upper right")
```
